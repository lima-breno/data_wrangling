##Importando um arquivo csv (com separação com vírgula)
 (é simples, n precisa especificar o separador!!)

```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
#Criando a variavel
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/superstore_data.csv'
```

```python
#Lendo os dados do url
dados_mercado = pd.read_csv(url)
```

```python
#verificando se o DataFrame está ok
dados_mercado.head()
```

## Importando um arquivo csv (separado por ponto e vírgula)
(precisamos acresecentar o separador!)

```python
#Criando ao variável
url_2 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/superstore_data_ponto_virgula.csv'
```

```python
#Lendo os dados do url
## ----> Aqui é necessário utilizar o sep = ';
dados_ponto_virgula = pd.read_csv(url_2, sep = ';')
```

```python
dados_ponto_virgula.head()
```

Formas diferentes de apresentar as primeiras linhas



```python
# utilizando o pd.read_csv ao inves do .head()
dados_primeiras_linhas = pd.read_csv(url, nrows = 5)
dados_primeiras_linhas
```

Forma de apresentar as colunas específicas

```python
#Utilizando o comando pd.read_csv (
    #especificando os nomes certinhos das colunas
dados_selecao = pd.read_csv(url, usecols = ['Id','Year_Birth', 'Income'])
dados_selecao
```

```python
# posso trazer os dados de acordo com o numero da coluna
  #Lembrando que python começa do zero
dados_selecao = pd.read_csv(url, usecols = [0,1,4])
dados_selecao
```

Salvando arquivos para csv

```python
dados_selecao.to_csv('clientes_mercado.csv')
```

```python
# É possível verificar que existe uma nova coluna Unnamed
  #é preciso ajustar na hora que for salvar
  # o comando pandas.DataFrame.to_csv por padra o acrescenta um
    #indice (coluna nos dados). Por padrão, ele fica como index = True
      #para poder remover este indice é necessário deixar index = False

#Salvando arquivo novamente
dados_selecao.to_csv('dados_mercado.csv', index = False)
```
