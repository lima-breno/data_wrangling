Case:

 App de delivery que guarda os dados dos cliente em uma planilha, e precisamos manipular alguns dados para melhorar a analise para futuros estudos

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/11. Manipulação de strings com Pandas/banco_clientes.csv')
df
```

- Padronizar os nomes dos clientes em letra maiúscula
- Padronizar a coluna de telefones para o formato XX XXXXX-XXXX
- Criar uma coluna apenas com os CEPs
-Ter uma coluna para o rua e outra para o número do endereço do cliente

#Padronizar os nomes em letra maiuscula

```python
df['nome'] = df['nome'].str.upper()
df
```

# Padronizar a coluna de telefones para o formato XX XXXXX-XXXX

```python
df['telefone'] = df.telefone.str.replace('(', '', regex=False).str.replace(')','', regex=False)
df
```

# Criar uma coluna apenas com CEPs

removendo o padrao do CEP com o metodo extract

```python
df['CEP'] = df.endereco.str.extract('(\d{5}-\d{3})')
df
```

```python
#Criada a coluna do CEP!!
# agora precisamos remover o dado do CEP da coluna endereço
#O comando str.strip() vai remover os espaços em branco antes e depoisi do dado! evitando problemas futuros

df['endereco'] = df.endereco.str.replace('(\d{5}-\d{3})', '',regex = True).str.strip()
df
```

```python
# CEP REMOVIDO DA COLUNA ENDEREÇO
```

# Ter uma coluna para a Rua e outra par ao numero do endereço do cliente

```python
# Separando os dados utilizando o separador , já disponivel no proprio dado

df.endereco.str.split(',',expand=True)
```

```python
# Ao fazer isso, é possível ver que alguns dados foram separados, outros não
 # isto ocorreu pois existiam dados que nao tinham a ,
  #podemos fazer isto com o regex!
  #O resultado de um split é um DF, e o replace é aplicado apenas em uma series.
    #-> entao quer dizer que preciso primeiro tirar a virgula para depois separar!
```

```python
df.endereco.str.replace(',','').str.split('\s(?=\d+$)', expand= True)
```

```python
# Salvando os dados em colunas separadas
df[['rua','numero']] = df.endereco.str.replace(',','').str.split('\s(?=\d+$)', expand= True, regex=True)
df
```

```python
# Removendo a coluna com a rua e numero agrupado
```

```python
df = df.drop(['endereco'], axis=1)
df
```
