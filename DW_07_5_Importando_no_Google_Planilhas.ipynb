Precisamos deixar essa planilha publica

Cada planilha do google tem um id, que começa depois do /d/ e vai até antes do /edit
No caso acima, o id é 1CPEc_erByjUr3tluewP7_AdAwJjmA_DT


```python
import pandas as pd
```

```python
#trazendo o id da planilha
sheet_id = '1CPEc_erByjUr3tluewP7_AdAwJjmA_DT'
```

```python
#preciso fazer um truquezinho para acessar a planilha
  # copiar ra parte inicial do link
  #trazer o id como o código
  # /gviz/tq -> é uma API do google que permite comunicação google colab com google planilhas
  # ?tqx=out:csv --> especifica que o formato retornado dessa API deve ser csv
  #&sheet --> vai retornar todos os dados da planilha, cabecalho, linhas, coluna, etc

url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

```python
#trazendo a url para uma variavel
dados_co2_sheets = pd.read_csv(url)
```

```python
#verificando se deu tudo certo
dados_co2_sheets.head()
```
