### Método groupby

```python
# Agrupamento de acordo com uma coluna específica, no noso caso é a Tipo
## O groupby funciona somente com uma outra fução junta, caso contrario vai dar a seguinte resposta
dados.groupby('Tipo')
```

```python
##Para poder corrigir, temos que acrescenctar uma função junto
## O metodo de agregaçaõ varia de acordo com  a versao do pandas
### numeric_only=True faz com que seja selecionado apenas nas colunas numerricas
geral = dados.groupby('Tipo').mean(numeric_only=True)
geral
```

```python
# Ele calcula a média de TODOS os dados numericos
```

```python
# para selecionar apenas a coluna dos Valor
  # Isso faz com que seja calculado APENAS a coluna Valor,
  ## isso faz com que n seja necessario colocar o 'numeric_only=True' pois ele ja vai selecionar apenas a coluna valor
  ### Deixar como numeric_only=True n vai dar problema, pq já foi selecionado a coluna valor
dados.groupby('Tipo')['Valor'].mean()
```

```python
# Transformando isso em um Dataframe
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# Plotando os dados em um grafico
##Colocando o DataFrame em uma variável
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

### código para poder plotar os dados
df_preco_tipo.plot(kind='barh', figsize =(14,10), color = 'red')
# kind -> Tipo de grafico que quero
# figsize -> Tamanho do gráfico
#color -> cor da barra
```

# **Aula 04 - Manipulando os dados**


```python
#IMportando a base de dados COMPLETA

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep = ';')
dados.head()
```

```python
#Criando uma coluna valor_por_mes no DataFrame dados

dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
#verificando se deu boa
dados.head()
```

```python
#criando valor anual
dados['Valor_por_ano'] = dados['Valor_por_mes']*12 + dados['IPTU']

#verificando se deu boa
dados.head()
```

# **Criando colunas Categóricas**
(Com palavras ou texto)

###Criando uma coluna Descrição

```python
#Como as colunas Tipo e Bairro são string, eu somando ambas vou concatenar as colunas

dados['Descricao'] = dados['Tipo'] + dados['Bairro']
dados.head()
```

```python
#Como personalizar as informações na coluna descrição

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
```

```python
#Acrescentando os quartos e vagas de garagem

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'] + ' quarto(s) ' \
       + ' e ' + dados['Vagas'] + ' vaga(s) de garagem.'
dados.head()
```

```python
#ERRO ACIMA ACONTECEU PQ: Só podemos concatenar str com str, nao posso concatenar str e int
 # Preciso converter dados da coluna quartos e vagas para str
  # PAra fazer isto, posso utilziar o comando astype(str) em vagas e quartos

dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + dados['Quartos'].astype(str) + ' quarto(s) ' + \
     ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

# Criando uma coluna binária

```python
#Criando uma coluna Possui_suite
  #Método apply vai aplicar uma função em um conjunto especifico de dados
    # lambda é uma função personalizavel, digitada em uma unica linha
    # O x representa cada linha do dataframe dados

dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
```

```python
#salvando esses dados em um arquivo

dados.to_csv('dados_completos_dev.csv', index = False, sep =';')
```
