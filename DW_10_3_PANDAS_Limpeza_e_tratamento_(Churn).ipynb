```python
# Esse comando vai retornar os indices contendo os outliers!
df_sem_nulo[outliers_index]['cliente.tempo_servico']
```

Substituindo valroes para os oulieres

```python
# fazendo uma copia do nosso DF sem alterar o DF principal

df_sem_out = df_sem_nulo.copy()
```

```python
#Identificando os outliers para ver se esta alinhado com os daos levantados anteriormente
df_sem_out[outliers_index]['cliente.tempo_servico']
```

```python
# esta igual! segue o baile!
# A coluna tempo_servico tem relacao direta com cobranca Total e cobranca mensal
# para verificarmos se foi feita a divisão correta podemos refazer o calculo da seguinte maneira

df_sem_out.loc[outliers_index, 'cliente.tempo_servico'] = np.ceil(
    df_sem_out.loc[outliers_index, 'conta.cobranca.Total'] /
    df_sem_out.loc[outliers_index, 'conta.cobranca.mensal']
)
```

# Removendo outliers

```python
#Precisamos passar de novo a função dos limites para poder remover os dados já corrigidos anteriormente
Q1 = df_sem_out['cliente.tempo_servico'].quantile(.25)
Q3= df_sem_out['cliente.tempo_servico'].quantile(.75)
IQR = Q3-Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR

outliers_index = (df_sem_out['cliente.tempo_servico'] < limite_inferior) | (df_sem_out['cliente.tempo_servico'] > limite_superior)
outliers_index
```

```python
# Tivemos uma series com booleana que sevirá como base para realizar o filtro
df_sem_out[outliers_index]
```

```python
# Retornou apenas 4 dados filtrados que ainda sao outlieres
# Para poder descartar essas amostras, fazemos o seguinte:
# Fazendo a atribuição do DF e removend os dados menos os outliers com o ~
df_sem_out = df_sem_out[~outliers_index]
df_sem_out
```

```python
# Temos os dados certinhos!
# Verificamos que as amostras vao até 7005 e temos 7002 linhas. Foram retiradas as amostras!

#Para reesetar o index
df_sem_out.reset_index(drop=True, inplace=True)
df_sem_out
```

### essa técnica tansforma variáveis categóricas em variáveis binárias!


```python
#Exemplo de como funciona o dummy
##Criando um DF de uma coluna com 3 cateorias distintas (a,b,c)
s = pd.Series(list('abca'))
s
```

```python
pd.get_dummies(s)
```

```python
# Retornou 4 amostras de 0 a 3 com 3 colunas correlacionando qual o dado que tinha na amostra
## EX: na amostra 0 (identificada na Serie(s) como a), ele traz True na coluna a mas false na coluna b,c
### a técnica dummy cria colunas e indica onde a amostra está localizada


#OBS:
  # a versao 2.0 do pandas retorna True e false, a versao 1.0 retorna 0 1
```

```python
#verificamos que temos colunas com dtype como object (que contem uma variavel nao categorica)

#APliCANDO A TECNICA DUMMIES NO DF

df_dummies = pd.get_dummies(df_sem_id, dtype=int).copy()
df_dummies.head()
```
