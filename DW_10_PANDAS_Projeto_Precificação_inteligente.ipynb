# Aula 03 - Manipulaçãõ de Dados textuais!

Tokenização = Divide um texto em unidades menores

EX: [um texto qlquer] -> ['um', 'texto', 'qlqer']



Manipulando para letras minusculas

```python
dados['descricao_local']
```

```python
#temos letras maiusculas e minusculars
  ##COMANDO STR -> vai fazer a leitura destes elementos das colunas como strings

dados['descricao_local'] = dados['descricao_local'].str.lower()
dados
```

```python
# Ficou tudo minusculo!
## MISSAO CUMPRIDA!
```

Removendo caracteres como / , .? ! @ ou demais caracteres especiais

-> Comando SRT

--> Comando REGEX: que irá encontrar todos os caracteres dos textos, sem encontrar letras, numeros, apostrofe e hifen (comuns na lingua inglesa)

```python
#Pegando um numero aleatorio para poder entender o que fazer
dados['descricao_local'][3169]
```

```python
# Fazendo a remoção dos caracteres indesejados

    # REGEX: o ^ faz com que os dados que virão a seguir dele ele nao vai selecioanr
      # isso aqui [aA-Z0-9\-\'] quer dizer para nao selecionar todas as letras e numeros, apostrofes e hifen
      #substituindo os que n queremos por espaço -> ' '

dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex = True)
```

```python
# REMOVENDO HIFEN FORA DE PALAVRAS COMPOSTAS
```

```python
#regex verifica se tem um caracter antes e depois do hifen, aí esse hifen será selecionado pelo regex
dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex = True)
```

```python
#hifen da linha 2 e 3816 se mantiveram! está certinho

# Fazendo a atribuição

dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex = True)
```

Tokenizaçao de strings

Método split -> divide os elementos textuais em uma lista. Ele pega o espaçoç em branco como elemento separador

```python
dados ['descricao_local'] = dados['descricao_local'].str.split()
```

```python
dados.head()
```

```python
#Coluna descrição_local está tokenizada, separada por vírgulas
```

```python
## Precisamos ajustaR a coluna comodidades!
  #Como: removendo o {} e as " "
    #Para isso, REGEX e REPLACE!
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex= True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
# PRONTO! A COLUNA COMODIDADE ESTA SEPARADA EM UMA LISTA SEPARADA POR VIRGULAS!
```
