Criando o banco de dados

```python
#Criando o banco de dados
#Utilizando um banco de dados do próprio colab
#memory indica que os dados sera armazenado na propria memoria
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
#Passando os dados para sql
#chamando o engine
#evitando a primeira coluna bugada
dados.to_sql('clientes',engine, index= False)
```

```python
#Vamos inspecionar o banco de dados com o inspect

inspector = inspect(engine)
```

```python
#printando o nome da tabela dentro do meu banco de dados
print(inspector.get_table_names())
```

```python
#Deu certo, o meu sql foi criado!
```

Lendo uma consulta SQL

```python
#Pesquisando as pessoas empregadas do banco de dados
#o * retorna todas as colunas da bas de dados
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
#lendo uma consulta em sql
empregados = pd.read_sql(query,engine)
```

```python
empregados
```

```python
# o con é utilizado para especificar o engine do sqlalchemy
empregados.to_sql( 'empregados', con=engine, index = False)
```

```python
#Fazendo uma leitura de tabelas
#OBS:
  #O read_sql é utilzado para ler uma consulta
  # O read_sql_table é utulizad para ler uma tabela !

pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

Deletando uma info (linha) de um banco de dados

```python
from string import whitespace
#Removendo uma linha do banco de dados

#Primeiro faz-se uma pesquisa
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
#para garantir que a conexão com o banco de dados seja encerrada corretamente apos a consulta:
with engine.connect() as conn:
  conn.execute(query)
```

```python
#Exemplo atualizando a escolaridade do final 508
query = 'UPDATE clientes SET Grau_escolaridade = "Ensino superior" WHERE ID_Cliente = 5008808'
#para garantir que a conexão com o banco de dados seja encerrada corretamente apos a consulta:
with engine.connect() as conn:
  conn.execute(query)
```
