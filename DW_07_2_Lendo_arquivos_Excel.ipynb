```python
dados_co2 = pd.read_excel(url)
```

```python
dados_co2.head()
```

# Parametros da função read_excel

Como saber se tenho mais abas/páginas dentro de uma planilha

```python
# O comando abaixo irá trazer como resposta os nomes das abas
pd.ExcelFile(url).sheet_names
```

Puxando uma aba/pagina específica do excel

```python
#Para puxar uma aba específica para o DataFrame posso usar o sheet_name da seguinte maneira
percapta = pd.read_excel(url,sheet_name = 'emissoes_percapita')
```

```python
percapta.head()
```

```python
fontes = pd.read_excel(url,sheet_name = 'fontes')
```

```python
fontes.head()
```

Como especificar as colunas que quero apresentar

```python
#Posso utilizar o comando usecols
  #no exemplo abaixo estou selecionando a aba emissoes_co2
  # Para identificar as colunas, é necessario trazer a info como o excel lê (coluna 1 é a A, etc)
  # O intervalo pode ser dado por ':'
intervalo = pd.read_excel(url,sheet_name = 'emissoes_C02', usecols = 'A:D')
```

```python
intervalo
```

Como especificar a quantide de linhas que eu quero

```python
# Posso utilizar o nrows!
intervalo_2 = pd.read_excel(url,sheet_name = 'emissoes_C02', usecols = 'A:D', nrows = 10)
```

```python
intervalo_2
```

Salvando arquivos em Excel

```python
#Salvando arquivo percapita
#Lembrar do parametro index
percapta.to_excel('co2_percapita.xlsx', index = False)
```
